```python
import pandas as pd

# Define column names
columns_names = ['target','id','date','flag','user','text']

# Load dataset without header row, apply custom column names
twitter_data = pd.read_csv("twitter.csv", encoding='ISO-8859-1', names=columns_names)

# Check shape and first rows
print(twitter_data.shape)
print(twitter_data.head())
```

```python
twitter_data.isnull().sum()
```

```python
twitter_data['target'].value_counts()
```

```python
twitter_data.replace({'target': {4: 1}}, inplace=True)
```

```python
twitter_data['target'].value_counts()
```

```python
from nltk.stem import PorterStemmer
port_stem=PorterStemmer()
```

```python
def stemming(content):
    stemmed_content=re.sub('[^a-zA-Z]',' ',content)
    stemmed_content=stemmed_content.lower()
    stemmed_content=stemmed_content.split()
    stemmed_content=[port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
    stemmed_content=' '.join(stemmed_content)
    return stemmed_content
```

```python
twitter_data['stemmed_content']=twitter_data['text'].apply(stemming)
```

```python
twitter_data.head()
```

```python
print(twitter_data['target'])
```

```python
#seperating the data and label
x=twitter_data['stemmed_content'].values
y=twitter_data['target'].values
```

```python
print(y)
```

```python
x_train.x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
#coverting the textual data to numerical data
```

```python
vectorizer=TfidVectorizer()
```

```python
x_train=vectorizer.fit_transform(x_train)
x_test=vectorizer.transform(x_test)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
model=LogisticRegression(max_iter=1000)
```

```python
model.fit(x_train,y_train)
```

```python
#accuracy score on training data
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(y_train,x_train_prediction)
```

```python
print('Accuracy_score on the training data:',training_data_accuracy)
```
